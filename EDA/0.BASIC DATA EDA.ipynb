## 월주차 column 추가
- 놀토와 놀일 등 첫 주차 둘째주차 휴일이 존재함.

```python
from math import ceil
def week_of_month(dt):
    
    first_day = dt.replace(day=1)
    dom = dt.day
    adjusted_dom = dom + first_day.weekday()
    # 주차 얻기
    week_of_month = int(ceil(adjusted_dom/7.0))
    # 첫 주차 세번째 주차 묶기 둘쨰주차 넷쨰 주차 묶기
    if (week_of_month == 1) or (week_of_month == 3):
        return 0
    elif (week_of_month == 2) or (week_of_month == 4):
        return 1
    else:
        return 2

# Apply the function to the 'date_time' column to create the 'week_of_month' column
train_df['week_of_month'] = train_df['date_time'].apply(week_of_month) 
test_df['week_of_month'] = test_df['date_time'].apply(week_of_month) 
```

### 3. 선형 회귀로 분석해보자.
- 일평균값으로 한번 linear regression의 R_square를 계산해보자.
- 매우 관련성이 높은 확인해 볼 수 있다.

```python
train_df = pd.read_csv('../data_file/fix_holiday_feature_train.csv')
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# linear regression 계수와 score return
def perform_linear_regression(X, y):
    reg = LinearRegression().fit(X, y)
    return reg.coef_[0], reg.intercept_, reg.score(X, y)
```

```python
data = train_df.copy()
data['date'] = pd.to_datetime(data['date_time']).dt.date
```

```python
# grouping
grouped_data = data[data['building_number'] == 1].groupby(['date', 'holiday']).agg({
    'temperature': 'mean',
    'power_consumption': 'mean'
}).reset_index()
# 쉬는날과 아닌 날 구분
grouped_holiday_data = grouped_data[grouped_data['holiday'] == 1]
grouped_non_holiday_data = grouped_data[grouped_data['holiday'] == 0]
# Grouped Holiday data
X_grouped_holiday = grouped_holiday_data[['temperature']]
y_grouped_holiday = grouped_holiday_data['power_consumption']
coef_grouped_holiday, intercept_grouped_holiday, score_grouped_holiday = perform_linear_regression(X_grouped_holiday, y_grouped_holiday)

# Grouped Non-Holiday data
X_grouped_non_holiday = grouped_non_holiday_data[['temperature']]
y_grouped_non_holiday = grouped_non_holiday_data['power_consumption']
coef_grouped_non_holiday, intercept_grouped_non_holiday, score_grouped_non_holiday = perform_linear_regression(X_grouped_non_holiday, y_grouped_non_holiday)

print('쉬는 날의 Rsquare 점수: ', score_grouped_holiday)

print('일하는 날의 Rsquare 점수: ',score_grouped_non_holiday)
```

## 5. 최대온도와 최대전력소모량이 나타나는 시간은 관계가 있을까?
- 관계는 분명 존재하는 것처럼 보입니다.
- 적게는 2시간 많게는 6시간도 차이나는 것으로 보이네요.
- 최대전력 소모량이 언제 발생하는지 찾는게 중요하다는 생각이 드네요.

```python
building_17_data = data[(data['building_number'] == 50) & (data['holiday'] == 0)]

building_17_data['date'] = pd.to_datetime(building_17_data['date_time']).dt.date

# 최고온도와 최대전력 소모량의 시간 찾기
hottest_times = building_17_data.loc[building_17_data.groupby('date')['temperature'].idxmax()]['date_time']
highest_power_times = building_17_data.loc[building_17_data.groupby('date')['power_consumption'].idxmax()]['date_time']

hottest_times.reset_index(drop=True, inplace=True)
highest_power_times.reset_index(drop=True, inplace=True)
```

```python
import matplotlib.pyplot as plt

# Extracting hour from the 'date_time' column for plotting
hottest_hours = pd.to_datetime(hottest_times).dt.hour
highest_power_hours = pd.to_datetime(highest_power_times).dt.hour

dates = building_17_data['date'].unique()

# Plotting the results
plt.figure(figsize=(14, 6))

plt.plot(dates, hottest_hours, label='Hottest Hour', marker='o', color='red')
plt.plot(dates, highest_power_hours, label='Highest Power Consumption Hour', marker='x', color='blue')

plt.xlabel('Date')
plt.ylabel('Hour of the Day')
plt.title('Hottest Hour vs. Highest Power Consumption Hour for Building 18')
plt.legend()
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()

plt.show()
```
