```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.signal import argrelextrema
import numpy as np
from datetime import datetime, timedelta
import warnings
```

```python
# warnings.simplefilter(action="ignore", category=SettingWithCopyWarning)
```

```python
df = pd.read_csv('../data_file/isolate_merge_df.csv')
```

```python
test_df = pd.read_csv('../data_file/fill_na_test.csv')
```

### holiday 37부터 42은 이상한 양상을 띈다. 기존 holiday feature를 극솟값에 대해서 0이되게 수정

```python
def find_holiday(df, i):
    building_df = df[df['building_number'] == i]
    building_df.loc[:, ['date_time']] = pd.to_datetime(building_df['date_time'], format='%Y-%m-%d %H:00:00')
    building_df.set_index('date_time', inplace=True)
    building_df_daily_max = building_df.resample('D').max()
    # 평균
    avg_pwr = (building_df_daily_max['power_consumption'].max() + building_df_daily_max['power_consumption'].min())/2
    
    # local minima 찾기
    comparator = np.less
    local_minima_indices = argrelextrema(building_df_daily_max['power_consumption'].values, comparator)
    local_minima = building_df_daily_max['power_consumption'].iloc[local_minima_indices]
    local_minima = local_minima[local_minima < avg_pwr]
    #print(local_minima)
    
    building_df.loc[:, 'holiday'] = 0
    
    for i in local_minima.index:
        print(i)
        building_df.loc[building_df.index[(building_df.index >= i) & (building_df.index  < (i + timedelta(days=1)))], 'holiday'] = 1
        
    return building_df
    
    
```

```python
find_holiday(df, 86)
```

### test set은 다 0으로 수정(휴일없음)

```python
for i in range(37, 43):
    test_df.loc[test_df['building_number'] == i, 'holiday'] = 0
```

### train set 수정

```python
for i in range(37, 43):
    building_df = find_holiday(df, i)
    df.loc[df['building_number'] == i, 'holiday'] = building_df['holiday'].values
```

### 5번 건물 수정

```python
df5 = df.loc[df['building_number'] == 5, :].copy()
df.loc[df['building_number'] == 5, 'holiday'] = df5.apply(lambda x : 1 if x['day']<4 else 0, axis = 1)
df.loc[('2022-06-17'<=df.date_time) & (df.date_time<'2022-06-18') & (df['building_number'] == 5), 'holiday'] = 1

test_df5 = test_df.loc[df['building_number'] == 5, :].copy()
test_df.loc[test_df['building_number'] == 5, 'holiday'] = test_df5.apply(lambda x : 1 if x['day']<4 else 0, axis = 1)
```

### 54번

```python
df54 = df.loc[df['building_number'] == 54, :].copy()
df.loc[df['building_number'] == 54, 'holiday'] = df54.apply(lambda x : 1 if x['day']<1 else 0, axis = 1)
df.loc[('2022-08-16'<=df.date_time) & (df.date_time<'2022-08-18') & (df['building_number'] == 54), 'holiday'] = 1

test_df54 = test_df.loc[df['building_number'] == 54, :].copy()
test_df.loc[test_df['building_number'] == 54, 'holiday'] = test_df54.apply(lambda x : 1 if x['day']<1 else 0, axis = 1)
```

### 86번

```python
building_df = find_holiday(df, 86)
building_df.reset_index(inplace=True)
building_df.loc[('2022-06-26'<=building_df.date_time) & (building_df.date_time<'2022-06-27'), 'holiday'] = 1
df.loc[df['building_number'] == 86, 'holiday'] = building_df['holiday'].values

test_df86 = test_df.loc[df['building_number'] == 86, :].copy()
test_df.loc[test_df['building_number'] == 86, 'holiday'] = test_df86.apply(lambda x : 1 if x['day']>5 else 0, axis = 1)
```

### 91번

```python
building_df = find_holiday(df, 91)
# 6월 12일 추가 6월 8일 삭제 7월 10일 추가
building_df.reset_index(inplace=True)
building_df.loc[('2022-06-12'<=building_df.date_time) & (building_df.date_time<'2022-06-13'), 'holiday'] = 1
building_df.loc[('2022-06-08'<=building_df.date_time) & (building_df.date_time<'2022-06-09'), 'holiday'] = 0
building_df.loc[('2022-07-10'<=building_df.date_time) & (building_df.date_time<'2022-07-11'), 'holiday'] = 1
# test_df는 일욜만 휴일

test_df91 = test_df.loc[df['building_number'] == 91, :].copy()
test_df.loc[test_df['building_number'] == 91, 'holiday'] = test_df91.apply(lambda x : 1 if x['day']>5 else 0, axis = 1)
```
