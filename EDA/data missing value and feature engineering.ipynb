```python
test_df.loc[train_df['rainfall'] > 0, 'rainfall'] = 1
test_df.loc[train_df['humidity'] == 100, 'rainfall'] = 1
test_df['rainfall_next_day'] = test_df['rainfall'].shift(24)

# Fill the NaN values in the 'rainfall_next_day' with 0
test_df['rainfall_next_day'] = test_df['rainfall_next_day'].fillna(0)
```

```python
train_df['rainfall_any'] = train_df.groupby(['day_of_year'])['rainfall_next_day'].transform(lambda x: 1 if x.sum() > 0 else 0)
test_df['rainfall_any'] = test_df.groupby(['day_of_year'])['rainfall_next_day'].transform(lambda x: 1 if x.sum() > 0 else 0)
```

```python
train_df['windspeed'] = train_df['windspeed'].interpolate(method='time')
train_df['humidity'] = train_df['humidity'].interpolate(method='time')
test_df['windspeed'] = test_df['windspeed'].interpolate(method='time')
test_df['humidity'] = test_df['humidity'].interpolate(method='time')
```

- cooling degree hours (CDH) 와 일평균 CDH 추가(냉방도일)
- 온도에서 24도를 뺀 값
- 24도부터 에어컨을 튼다고 함. 이후 얼마나 사용할지 가정.

```python
def CDH(xs):
    ys = []
    for i in range(len(xs)):
        if i < 11:
            ys.append(np.sum(xs[:(i+1)]-24))
        else:
            ys.append(np.sum(xs[(i-11):(i+1)]-24))
    return np.array(ys)
```

```python
cdhs = np.array([])
for num in range(1,101):
    temp = train_df[train_df['building_number'] == num]
    cdh = CDH(temp['temperature'].values)
    cdhs = np.concatenate([cdhs, cdh])
train_df['cdh'] = cdhs
```

```python
cdhs = np.array([])
for num in range(1,101):
    temp = test_df[test_df['building_number'] == num]
    cdh = CDH(temp['temperature'].values)
    cdhs = np.concatenate([cdhs, cdh])
test_df['cdh'] = cdhs
```
