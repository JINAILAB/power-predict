```python
train_df = pd.read_csv('../data_file/basic_train.csv')
test_df = pd.read_csv('../data_file/basic_test.csv')
```

- 결측치 처리 풍속 , 습도 interpolate
- capacity들은 -을 0으로 변환
- 비 drop
- solar radiation, sunshine drop

```python
train_df['ddt'] = pd.to_datetime(train_df['date_time'], format='%Y%m%d %H')
train_df.set_index('ddt', inplace=True)
test_df['ddt'] = pd.to_datetime(test_df['date_time'], format='%Y%m%d %H')
test_df.set_index('ddt', inplace=True)
```

```python
train_df['windspeed'] = train_df['windspeed'].interpolate(method='time')
train_df['humidity'] = train_df['humidity'].interpolate(method='time')
test_df['windspeed'] = test_df['windspeed'].interpolate(method='time')
test_df['humidity'] = test_df['humidity'].interpolate(method='time')
```

- DI(discomfort index) 추가 
- 0.81×T+0.01×H×(0.99×T−14.3)+46.3

```python
train_df['discomfort_index'] = 0.81 * train_df['temperature'] + 0.01 * train_df['humidity'] * (0.99 * train_df['temperature'] - 14.3) + 46.3
test_df['discomfort_index'] = 0.81 * test_df['temperature'] + 0.01 * test_df['humidity'] * (0.99 * test_df['temperature'] - 14.3) + 46.3
```

```python
train_df['solar_power_capacity'][train_df['solar_power_capacity'] == '-'] = 0
test_df['solar_power_capacity'][test_df['solar_power_capacity'] == '-'] = 0
```

```python
train_df.drop(['sunshine', 'solar_radiation', 'rainfall', 'ess_capacity', 'pcs_capacity'], axis=1, inplace=True)
test_df.drop(['rainfall', 'ess_capacity', 'pcs_capacity'], axis=1, inplace=True)
```

- cooling degree hours (CDH) 와 일평균 CDH 추가(냉방도일)
- 온도에서 24도를 뺀 값
- 24도부터 에어컨을 튼다고 함. 이후 얼마나 사용할지 가정.

```python
def CDH(xs):
    ys = []
    for i in range(len(xs)):
        if i < 11:
            ys.append(np.sum(xs[:(i+1)]-24))
        else:
            ys.append(np.sum(xs[(i-11):(i+1)]-24))
    return np.array(ys)
```

```python
cdhs = np.array([])
for num in range(1,101):
    temp = train_df[train_df['building_number'] == num]
    cdh = CDH(temp['temperature'].values)
    cdhs = np.concatenate([cdhs, cdh])
train_df['cdh'] = cdhs
```

```python
cdhs = np.array([])
for num in range(1,101):
    temp = test_df[test_df['building_number'] == num]
    cdh = CDH(temp['temperature'].values)
    cdhs = np.concatenate([cdhs, cdh])
test_df['cdh'] = cdhs
```

- cdh와 total area를 곱해서 전력량 예측 산출

```python
train_df['total_area_x_cdh'] = train_df['total_area'] * train_df['cdh'] * 0.00001
train_df['cooling_area_x_cdh'] = train_df['cooling_area'] * train_df['cdh'] * 0.00001
test_df['total_area_x_cdh'] = test_df['total_area'] * test_df['cdh'] * 0.00001
test_df['cooling_area_x_cdh'] = test_df['cooling_area'] * test_df['cdh'] * 0.00001
```

```python
train_df['date_time'] = pd.to_datetime(train_df['date_time'], format='%Y%m%d %H')
train_df['hour'] = train_df['date_time'].dt.hour
train_df['week'] = train_df['date_time'].dt.isocalendar().week
```

```python
test_df['date_time'] = pd.to_datetime(test_df['date_time'], format='%Y%m%d %H')
test_df['hour'] = test_df['date_time'].dt.hour
test_df['week'] = test_df['date_time'].dt.isocalendar().week
```

```python
train_df['sin_time'] = np.sin(2*np.pi*train_df.hour/24)
train_df['cos_time'] = np.cos(2*np.pi*train_df.hour/24)
```

```python
test_df['sin_time'] = np.sin(2*np.pi*test_df.hour/24)
test_df['cos_time'] = np.cos(2*np.pi*test_df.hour/24)
```

```python
train_df['holiday'] = train_df.apply(lambda x : 0 if x['day']<5 else 1, axis = 1)
train_df['holiday'] = train_df.apply(lambda x : 2 if x['day']==6 else x['holiday'], axis = 1)
test_df['holiday'] = test_df.apply(lambda x : 0 if x['day']<5 else 1, axis = 1)
test_df['holiday'] = test_df.apply(lambda x : 2 if x['day']==6 else x['holiday'], axis = 1)
train_df.loc[('2022-06-01'<=train_df.date_time)&(train_df.date_time<'2022-06-02'), 'holiday'] = 1
train_df.loc[('2022-06-06'<=train_df.date_time)&(train_df.date_time<'2022-06-07'), 'holiday'] = 1
train_df.loc[('2022-08-15'<=train_df.date_time)&(train_df.date_time<'2022-08-16'), 'holiday'] = 1
```

```python
train_df.drop(['temperature', 'windspeed', 'humidity','month', 'year', 'hour','day_of_year', 'hour_of_year', 'day_of_week'], axis=1, inplace=True)
```
